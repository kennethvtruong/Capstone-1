```python
income_data[['County', 'Median household income','State Code','State']]
income_data['CountyState'] = income_data['County'] + income_data['State Code']
income_data
```

```python
merged_data_1 = pd.merge(income_data, rent_1b_2015, on='CountyState')
merged_data_1.drop(['County_y','State Code_y'], axis=1, inplace=True)
merged_data_1.columns = ['County-State','County','Population','Median household income',
                       'State Code','State', 'CountyState','1B Price']
merged_data_1
```

```python
merged_data_1['1B Price'] = 12*merged_data_1['1B Price']
merged_data_1['1B Income-Price Difference'] = merged_data_1['Median household income'] - merged_data_1['1B Price']
merged_data_1
```

```python
merged_data_2 = pd.merge(income_data, rent_2b_2015, on='CountyState')
merged_data_2.drop(['County_y','State Code_y'], axis=1, inplace=True)
merged_data_2.columns = ['County-State','County','Population','Median household income',
                       'State Code','State', 'CountyState','2B Price']
```

```python
merged_data_2['2B Price'] = 12 * merged_data_2['2B Price']
merged_data_2['2B Income-Price Difference'] = merged_data_2['Median household income'] - merged_data_2['2B Price']
merged_data_2
```

```python
merged_data_3 = pd.merge(income_data, rent_3b_2015, on='CountyState')
merged_data_3.drop(['County_y','State Code_y'], axis=1, inplace=True)
merged_data_3.columns = ['County-State','County','Population','Median household income',
                       'State Code','State', 'CountyState','3B Price']
```

```python
merged_data_3['3B Price'] = 12 * merged_data_3['3B Price']
merged_data_3['3B Income-Price Difference'] = merged_data_3['Median household income'] - merged_data_3['3B Price']
merged_data_3
```

```python
merged_df = pd.merge(merged_data_1, merged_data_2, on='CountyState', how='outer')

merged_df['County-State_y'].fillna(merged_df['County-State_x'], inplace=True)
merged_df['County_y'].fillna(merged_df.County_x, inplace=True)
merged_df['Population_y'].fillna(merged_df.Population_x, inplace=True)
merged_df['Median household income_y'].fillna(merged_df['Median household income_x'], inplace=True)
merged_df['State Code_y'].fillna(merged_df['State Code_x'], inplace=True)
merged_df['State_y'].fillna(merged_df['State_x'], inplace=True)

merged_df.drop(['State Code_x','State_x','Population_x','County-State_x',
               'County_x','Median household income_x'], axis=1, inplace=True)
merged_df.columns = ['CountyState', '1B Price','1B Income-Price Difference','County-State','County','Population',
                  'Median household income','State Code','State','2B Price','2B Income-Price Difference']
merged_df_1b2b = merged_df[['County-State','County', 'State Code', 'Population', '1B Price', 
                           '1B Income-Price Difference', '2B Price', '2B Income-Price Difference']]
merged_df_1b2b
```

```python
merged_all = pd.merge(merged_df, merged_data_3, on='CountyState', how='outer')

merged_all['County-State_y'].fillna(merged_all['County-State_x'], inplace=True)
merged_all['County_y'].fillna(merged_all.County_x, inplace=True)
merged_all['Population_y'].fillna(merged_all.Population_x, inplace=True)
merged_all['Median household income_y'].fillna(merged_all['Median household income_x'], inplace=True)
merged_all['State Code_y'].fillna(merged_all['State Code_x'], inplace=True)
merged_all['State_y'].fillna(merged_all['State_x'], inplace=True)

merged_all.drop(['State Code_x','State_x','Population_x','County-State_x','County_x','Median household income_x'], axis=1, inplace=True)
merged_all.columns = ['CountyState', '1B Price','1B Income-Price Difference','2B Price','2B Income-Price Difference',
                      'County-State','County','Population','Median household income','State Code','State', '3B Price', 
                     '3B Income-Price Difference']

merged_all[190:210]
```

```python
merged_all['1B Income-Price Difference-norm'] = merged_all['Median household income'] - merged_all['1B Price']
merged_all['2B Income-Price Difference-norm'] = merged_all['Median household income'] - (merged_all['2B Price'] / 2)
merged_all['3B Income-Price Difference-norm'] = merged_all['Median household income'] - (merged_all['3B Price'] / 3)
merged_all
```
